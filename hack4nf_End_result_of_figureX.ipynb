```python

import matplotlib as mpl

def set_pub_style():
    mpl.rcParams.update({

        "font.size": 14,
        "axes.titlesize": 16,
        "axes.labelsize": 16,
        "xtick.labelsize": 12,
        "ytick.labelsize": 12,



        "axes.linewidth": 1.6,
        "lines.linewidth": 1.8,
        "xtick.major.width": 1.4,
        "ytick.major.width": 1.4,
        "xtick.major.size": 6,
        "ytick.major.size": 6,



        "axes.grid": False,
        "grid.linewidth": 0.6,
        "grid.alpha": 0.25,


        "savefig.dpi": 300,
        "figure.dpi": 120,


        "figure.constrained_layout.use": True,
    })
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def plot_figureX(df, title, out_png, y_min=None):
    x = np.arange(len(df))
    w = 0.38

    plt.figure(figsize=(11, 4.8))
    plt.bar(x - w/2, df["s_wt"], width=w, label="wild type")
    plt.bar(x + w/2, df["s_mut"], width=w, label="mutant")

    plt.xticks(x, df["compound"], rotation=90, fontsize=8)
    plt.ylabel("S′ value")
    plt.title(title)
    plt.legend(frameon=False)

    if y_min is not None:
        plt.ylim(bottom=y_min)


    plt.tight_layout()
    plt.show()


# Panel A
dfA = pd.read_sql("""
SELECT
    name AS compound,
    ref_pooled_s_prime  AS S_wt,
    test_pooled_s_prime AS S_mut,
    delta_s_prime
FROM public.fnl_sprime_pooled_delta_sprime
WHERE
    tissue = 'PLEURA'
    AND ref_pooled_s_prime  > 2
    AND test_pooled_s_prime > 2
    AND delta_s_prime BETWEEN -2 AND 0
ORDER BY ref_pooled_s_prime ASC;
""", conn)

plot_figureX(
    dfA,
    "Figure X. Selection of compound candidates to target wild type and mutant cells",
    "FigureX_panelA.png",
    y_min=0
)

# Panel B
dfB = pd.read_sql("""
SELECT
    name AS compound,
    ref_pooled_s_prime  AS S_wt,
    test_pooled_s_prime AS S_mut,
    delta_s_prime
FROM public.fnl_sprime_pooled_delta_sprime
WHERE
    tissue = 'PLEURA'
    AND ref_pooled_s_prime  > -0.5
    AND test_pooled_s_prime > 2
    AND delta_s_prime BETWEEN -2 AND 0
ORDER BY ref_pooled_s_prime ASC;
""", conn)

plot_figureX(
    dfB,
    "Figure X. Selection of compound candidates to preferentially target mutant cells",
    "FigureX_panelB.png"
)
```
